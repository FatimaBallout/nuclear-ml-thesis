```python
import os
import pandas as pd
import re

def extract_cell_data_v3(lines):
    # Define a pattern to identify cell data lines
    cell_data_pattern = re.compile(r"^\s*([\d\.E+-]+)\s+([\d\.E+-]+)\s+([\d\.E+-]+)")

    data = []
    current_cell = None
    capture = False  # Flag to start capturing data

    for line in lines:
        if "cell" in line and re.search(r"\d+", line):
            current_cell = int(re.search(r"\d+", line).group())
            capture = True
        elif "energy" in line and capture:
            continue
        elif current_cell is not None and capture:
            match = cell_data_pattern.match(line)
            if match:
                energy, value1, value2 = map(float, match.groups())
                data.append([current_cell, energy, value1, value2])
            else:
                # Reset current cell and capture flag if line doesn't match the pattern
                current_cell = None
                capture = False

    df = pd.DataFrame(data, columns=["Cell", "Energy", "Flux", "Error"])
    return df

def extract_cell_data_from_file(file_path, mode='r'):
    with open(file_path, mode) as file:
        lines = file.readlines()
    return extract_cell_data_v3(lines)

# Define your directory path here
directory_path = r"/Users/fatimaballout/downloads"

# Define your file path here
file_path = os.path.join(directory_path, 'zD00401N_MODab_L0500600_5y_R_120_D05C9DEC485BD5A13F91D658208745C7FE43613FF1D170820D35C4FE42D5ED091BDC9D051.o')

# Process the specific file
df_name = os.path.splitext(os.path.basename(file_path))[0]  # Extract the filename without extension
dataframe = extract_cell_data_from_file(file_path)

# Extract unique energy values
unique_energy_values = dataframe["Energy"].unique()

# Create a new DataFrame with unique energy values
energy_df = pd.DataFrame({"Energy": unique_energy_values})

# Merge the energy DataFrame with the original DataFrame to associate flux and error values
result_df = energy_df.merge(dataframe, on="Energy", how="left")

# Now, result_df contains a clear representation of the data with unique energy values.
print(result_df)
```
